```python
import pandas as pd
import numpy as np
import transformers
from transformers import AutoModelForSequenceClassification
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments, DistilBertTokenizerFast, AutoTokenizer, AutoModelForSequenceClassification, default_data_collator
from datasets import load_dataset, load_metric
import torch
```

```python
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=2)
mnli_model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)
stsb_model = AutoModelForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=1)
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
)
```

```python
def compute_metrics(eval_pred, task):
  logits, labels = eval_pred
  if task == 'stsb':
        predictions = logits.squeeze()
        labels = labels.squeeze()
        return {"pearson": np.corrcoef(predictions, labels)[0, 1]}
  else:
        metric = load_metric('glue', task)
        predictions = np.argmax(logits, axis=1)
        return metric.compute(predictions=predictions, references=labels)
```

```python
def tokenize_function(examples, task):
  if task in ['cola', 'sst2']:
    result = tokenizer(examples['sentence'], truncation=True, padding='max_length')
  elif task in ['mrpc', 'rte', 'wnli']:
    result = tokenizer(examples['sentence1'], examples['sentence2'], truncation=True, padding='max_length')
  elif task in ['stsb']:
    result = tokenizer(examples['sentence1'], examples['sentence2'], truncation=True, padding='max_length')
    result['label'] = [float(label) for label in examples['label']]
  elif task in ['mnli']:
    result = tokenizer(examples['premise'], examples['hypothesis'], truncation=True, padding='max_length')
  elif task in ['qqp']:
    result = tokenizer(examples['question1'], examples['question2'], truncation=True, padding='max_length')
  elif task in ['qnli']:
    result = tokenizer(examples['question'], examples['sentence'], truncation=True, padding='max_length')
  else:
    raise ValueError('Unknown task')

  return result
```

```python
result = {}
dataset = load_dataset('glue', 'cola')
tokenized_dataset = dataset.map(lambda examples: tokenize_function(examples, 'cola'), batched=True)
train_dataset = tokenized_dataset['train'].shuffle(seed=42).select(range(min(5000, len(dataset['train']))))
eval_dataset = tokenized_dataset['validation'].shuffle(seed=42).select(range(min(5000, len(dataset['validation']))))

metric = load_metric('glue', 'cola')
trainer = Trainer(
  model=model,
  args=training_args,
  train_dataset=train_dataset,
  eval_dataset=eval_dataset,
  compute_metrics=lambda eval_pred: compute_metrics(eval_pred, 'cola')
)
trainer.train()
eval_result= trainer.evaluate()
result["cola"] = eval_result
```

```python
from datasets import Dataset
# Preprocess emotion dataset
emotion = load_dataset('emotion')
data = []
positive = [1,2,5]
negative = [3,4,6]
for example in emotion['train']:
  sentence = example['text']
  label = 1 if example['label'] in positive else 0
  data.append({'sentence': sentence, 'label': label})
df = pd.DataFrame(data).sample(n=5000, random_state=42)
emotion_sst2_format = Dataset.from_pandas(df)
```
